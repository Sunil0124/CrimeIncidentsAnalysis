```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
# Load the dataset
file_name = "Crime Incidents data.csv"
crm_data = pd.read_csv(file_name, low_memory=False)

# Display the first few rows
print(crm_data.head())
```

```python
crm_data = crm_data.dropna(axis=1, how='all')
crm_data = crm_data. dropna ( )
print (crm_data.head ( ))
```

```python
print(crm_data. info ())
crm_data['INCIDENT_NUMBER'] = crm_data['INCIDENT_NUMBER'].astype(str)
# Verify cleaned data
print(crm_data. info ())
```

```python
plt.figure(figsize=(10, 6))
crm_data['YEAR'].hist(bins=range(2000, 2025), color='red', edgecolor= 'black')
plt.title( 'Crime Incidents per Year')
plt.xlabel('Year')
plt.ylabel( 'Number of Incidents')
plt.grid(axis='y')
plt. show()
```

```python
plt.figure(figsize=(10, 6))
crm_data ['HOUR'].hist(bins=24, color='brown', edgecolor='black')
plt.title('Crime Incidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Incidents')
plt.grid(axis='y')
plt.show( )
```

```python
plt.figure(figsize=(12, 6))
crm_data['DISTRICT'].value_counts().plot(kind='bar', color='gold')
plt.title('Crime Incidents by District')
plt.xlabel('District')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show( )
```

```python
plt.figure(figsize=(12, 6))
heatmap_data = crm_data.pivot_table(index='DAY_OF_WEEK', columns='DISTRICT', values='INCIDENT_NUMBER', aggfunc='count')
sb.heatmap(heatmap_data, cmap='Purples', annot=True, fmt='g')
plt.title('Crime Incidents by Day of the Week and District')
plt.xlabel('District')
plt.ylabel('Day of the Week')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
monthly_trend = crm_data.groupby(['YEAR', 'MONTH']).size().reset_index(name='Incidents')
sb.lineplot(x='MONTH', y='Incidents', hue='YEAR', data=monthly_trend, marker='o')
plt.title('Monthly Crime Incidents Over Time')
plt.xlabel ('Month' )
plt.ylabel('Number of Incidents')
plt.show( )
```

```python
Shooting_df = crm_data[crm_data['SHOOTING'] == 1]
plt.figure(figsize=(10,6))
plt.scatter(Shooting_df['HOUR'], Shooting_df['OFFENSE_CODE'], alpha=0.5, color='black', s=10)
plt.title("Hour vs Shooting Incidents")
plt.xlabel('Hour of the Day')
plt.ylabel('Offense Code (Shooting)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_offenses = crm_data['OFFENSE_DESCRIPTION'].value_counts().head(5).index
offense_data = crm_data[crm_data['OFFENSE_DESCRIPTION'].isin(top_offenses)]
offense_counts = offense_data.groupby(['DISTRICT', 'OFFENSE_DESCRIPTION']).size().unstack().fillna(0)
# Define a professional color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
offense_counts.plot(kind='bar', stacked=True, color=colors)
plt.title('Top 5 Offense Types by District')
plt.xlabel('District')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = crm_data.pivot_table(index='DAY_OF_WEEK', columns='HOUR', values='INCIDENT_NUMBER', aggfunc='count')
plt. figure(figsize=(12, 6))
sb. heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Crime Heatmap by Day and Hour')
plt.xlabel ('Hour')
plt.ylabel('Day of the Week')
plt.show( )
```
